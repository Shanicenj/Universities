```python
address1 = universities['address']
type(address1), address1
```

```python
address1 = address1.str.strip()
address1
```

```python
address1 = address1.str.split(',', expand=True)
address1
```

```python
state_zip = address1[2].str.rsplit(' ', expand=True)
address1['street'] = state_zip[0]
address1['city_state'] = state_zip[1]
address1 = address1.rename(columns={0:'street', 1: 'city', 2:'state_zip'})
address1
```

```python
address2 = address1['state_zip'].str.rsplit(' ',1, expand=True)
address2 = address2.rename(columns={0:'state', 1: 'zip'})
address2
```

```python
data1 = pd.concat([universities,address1, address2], axis=1)
data1
```

```python
data1.drop(columns =['address', 'Website', 'street', 3, 4, 5, 'state_zip', 'city_state'], axis=1) 
```
